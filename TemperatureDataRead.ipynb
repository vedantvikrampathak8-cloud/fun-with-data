```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Delhi.csv")
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

plt.figure(figsize=(10,5))
plt.plot(df['date'], df['temperature_2m'], color='orange', linewidth=2)
plt.title("Temperature Trend in Delhi")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()

df['month'] = df['date'].dt.month
monthly_avg = df.groupby('month')['temperature_2m'].mean()
plt.figure(figsize=(8,4))
monthly_avg.plot(kind='bar', color='skyblue')
plt.title("Average Monthly Temperature in Delhi")
plt.xlabel("Month")
plt.ylabel("Average Temperature (°C)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.heatmap(df[['temperature_2m','relative_humidity_2m','pressure_msl','wind_speed_10m']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Weather Parameters")
plt.tight_layout()
plt.show()
```
